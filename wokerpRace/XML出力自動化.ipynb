```python
# Google spreadsheetsの接続
from google.colab import auth
auth.authenticate_user()
import gspread
from google.auth import default
creds, _ = default()
gc = gspread.authorize(creds)

# googleスプレッドシートの[xml]シートを取得
filename = "testplay"
ss = gc.open(filename)
ws = ss.worksheet('cardpool')
wsxml = ss.worksheet('xml')

for i in range(207):

  if i < 156:
    continue

  level = str(ws.cell(i+2, 2).value) # B列の2行目から取得
  if level == '1':
    foldername = "level1/"
  elif level == '2':
    foldername = "level2/"
  elif level == '3':
    foldername = "level3/"

  reverse = str(ws.cell(i+2, 18).value) # R列の2行目から取得
  if reverse == '1':
    uramenhensu = 4
  elif reverse == '2':
    uramenhensu = 5
  elif reverse == '3':
    uramenhensu = 6
  elif reverse == '12':
    uramenhensu = 7
  elif reverse == '13':
    uramenhensu = 8
  elif reverse == '23':
    uramenhensu = 9
  elif reverse == '11':
    uramenhensu = 10
  elif reverse == '22':
    uramenhensu = 11
  elif reverse == '33':
    uramenhensu = 12

  xmlstr_1 = str(wsxml.cell(2, 1).value) # A列の2行目を取得
  xmlstr_1 = xmlstr_1 + foldername + str(i) + ".png</data>" # ファイルパスを記載
  xmlstr_2 = str(wsxml.cell(uramenhensu, 1).value) # A列の4-12行目を取得

  print(xmlstr_1 + xmlstr_2)
```
